# Number of processed papers

```python
N = 3
ind = np.arange(N)
num = [1620048, 1087086, 230364]

rects = plt.bar(ind, num, width = 0.35, color='b')

labels = ['Downloaded', 'Parsed', 'Processed']
plt.ylabel('Number of Papers')
plt.xticks(ind + 0.35 / 2, labels, rotation = 30)
plt.title('Number of Processed Papers')

plt.savefig('numPapers.png', bbox_inches='tight')

plt.show()
```

# Communication Pie Chart

```python
import matplotlib.colors as colors
import matplotlib.cm as cmx

labels = 'eMail', 'Meetings','Whatsapp', 'Git', 'Slack'
sizes = [5, 60, 10, 20, 5]
explode = (0, 0.1, 0, 0,0)

cmap = plt.cm.Blues
colors = cmap(np.linspace(0., 1., len(labels)))

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90, colors = colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


plt.savefig('communication.png', bbox_inches='tight')
plt.show()
```

# Learning Curve

```python
def f(x): 
    return np.exp(x) -1

xvec = np.linspace(0,3)
y  = f(xvec)
plt.plot(xvec, y,lw = 3)
labels = ['Week 1-3', 'Week 4-7', 'Week 8-10','Week 11-12']
lablesy = ['Beginner', 'Intermediate', 'Expert']
#plt.title('Learning Curve')
xlab = [0,1,2,3]
plt.xticks(xlab, labels, rotation = 30)
plt.yticks([0,10,19],lablesy)


plt.fill_between(xvec, 0, y)

plt.savefig('learning.png', bbox_inches='tight')

plt.show()
```
